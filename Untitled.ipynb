```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10**6

range_Li = np.random.randn(int(N))*.8+3;
range_alpha = np.random.randn(int(N))*1.4+6.5;

ranges = np.append(range_Li[range_Li > 0],range_alpha[range_alpha > 0])

theta = np.random.rand(len(ranges))*180.;

x = np.cos(np.deg2rad(theta))*ranges/2;

plt.figure()
plt.hist(range_Li, bins=1000,color='b',alpha = .7);
plt.hist(range_alpha, bins=1000,color='r',alpha = .7);

plt.xlim([0,12]);
plt.title('Secondary particle range approximation')
plt.xlabel('range (mm)')
plt.legend(('Li7','alpha'))
plt.savefig('images_bin/range', dpi=300)


plt.figure()
plt.hist(x, bins=1000);
plt.xlabel('x (mm)');
plt.savefig('images_bin/d', dpi=300)
```
