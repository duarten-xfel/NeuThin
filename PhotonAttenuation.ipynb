```python
# IMPORTS
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,.2,500) #cm

# mu_rho = 342.2 # P10 (cm2/g)
# rho = 1.56e-3 # P10 (g/cm3)

mu_rho = 4.568 # Bi2O3 (cm2/g)
rho = 8.9 # Bi2O3 (g/cm3)


# mu_rho = 121.0 # Al (cm2/g)
# rho = 2.7 # Al (g/cm3)

mu = mu_rho*rho

frac_I = np.exp(-x*mu)

plt.figure()
plt.plot(x,frac_I);
plt.grid()
plt.xlim([0,max(x)]); plt.xlabel('x (cm)');

plt.ylim([0,1]); plt.ylabel('$I/I_0$');

# plt.xscale('log')
```
