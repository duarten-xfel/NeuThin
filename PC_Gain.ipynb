```python
# IMPORTS
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
##### Neutron Source ######


## Constants
e = -1.6e-19 # C (electron charge)
E = 2.79e6 # eV (incident radiation)    --> For neutron source

## Detector parameters

p = 1 # atm (gas pressure)
a = 50e-4/2 # cm (anode radius) 
b = 4.8/2 # cm (cathode radius) 

preamp_gain = 1000 # mV/pC

## Gas parameters (Ar-CO2)
W = 26
deltaV = 23.6 #V (from P10)
K = 4.8e4 # V/(cm*atm) (from P10)
########


HV = np.linspace(1400,2400,25) # V (applied HV)

logM = HV/np.log(b/a) * np.log(2)/deltaV * np.log(HV/(K*p*a*np.log(b/a)))
M = np.exp(logM);
n0 = E/W
Q = n0*e*M;
    
V_output = -1*Q*preamp_gain*1e9 # V    


plt.figure()
plt.plot(HV,M,'k--'); plt.yscale('Log');
plt.xlabel('HV (V)'); plt.ylabel('Gas Gain');

plt.figure()
plt.plot(HV,V_output,'--'); plt.yscale('Log');
plt.xlabel('HV (V)'); plt.ylabel('Max preamp output V (V)');
```
